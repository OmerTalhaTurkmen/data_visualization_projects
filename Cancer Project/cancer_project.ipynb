```python
# import libraries 
import pandas as pd # Import Pandas for data manipulation using dataframes
import numpy as np # Import Numpy for data statistical analysis 
import matplotlib.pyplot as plt # Import matplotlib for data visualisation
import seaborn as sns # Statistical data visualization
```

```python
# Import Cancer data 
cancer_df = pd.read_csv('cancer.csv')
```

```python
cancer_df.columns
```

```python
cancer_df.target.value_counts()
```

```python
plt.figure(figsize=(10,8))

plt.figure(figsize=(10, 8))

mylabels = ["Ä°yi Huylu", "KÃ¶tÃ¼ Huylu"]
custom_palette = sns.color_palette(["pink", "darkblue"])


plt.pie(cancer_df['target'].value_counts(), labels=mylabels,
        colors=custom_palette,
        autopct="%.1f%%",
        startangle=90,
        shadow=True,
        pctdistance=0.3,
        textprops={'color': 'white', 'fontsize': 15}
        )

plt.title('Target')

plt.legend(loc=(0))
plt.axis('equal')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax= sns.countplot(data=cancer_df, x= "target")

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.1,p.get_height()+1))
```

mean radius, mean perimeter, mean area, worst radius, worst perimeter, worst area gibi Ã¶zellikler kanser hÃ¼crelerinin boyutunu ifade eder. 
Bu Ã¶zelliklerin kanser hÃ¼crelerinin bÃ¼yÃ¼klÃ¼ÄŸÃ¼ Ã¼zerindeki etkilerini inceleyebilirsiniz.


```python

custom_palette = sns.color_palette(["darkblue", "pink"])

fig, ax = plt.subplots(2, 3, figsize=(25, 15))

sns.scatterplot(x="mean area", y="worst area", data=cancer_df, hue=cancer_df["target"], ax=ax[0, 0], palette=custom_palette)
ax[0, 0].set_xlabel("Mean Area", fontsize=14)
ax[0, 0].set_ylabel("Worst Area", fontsize=14)

sns.scatterplot(x="mean perimeter", y="worst perimeter", data=cancer_df, hue=cancer_df["target"], ax=ax[0, 1], palette=custom_palette)
ax[0, 1].set_xlabel("Mean Perimeter", fontsize=14)
ax[0, 1].set_ylabel("Worst Perimeter", fontsize=14)

sns.scatterplot(x="mean radius", y="worst radius", data=cancer_df, hue=cancer_df["target"], ax=ax[0, 2], palette=custom_palette)
ax[0, 2].set_xlabel("Mean Radius", fontsize=14)
ax[0, 2].set_ylabel("Worst Radius", fontsize=14)

sns.scatterplot(x="mean area", y="mean perimeter", data=cancer_df, hue=cancer_df["target"], ax=ax[1, 0], palette=custom_palette)
ax[1, 0].set_xlabel("Mean Area", fontsize=14)
ax[1, 0].set_ylabel("Mean Perimeter", fontsize=14)

sns.scatterplot(x="mean area", y="mean radius", data=cancer_df, hue=cancer_df["target"], ax=ax[1, 1], palette=custom_palette)
ax[1, 1].set_xlabel("Mean Area", fontsize=14)
ax[1, 1].set_ylabel("Mean Radius", fontsize=14)

sns.scatterplot(x="mean area", y="mean concavity", data=cancer_df, hue=cancer_df["target"], ax=ax[1, 2], palette=custom_palette)
ax[1, 2].set_xlabel("Mean Area", fontsize=14)
ax[1, 2].set_ylabel("Mean Concavity", fontsize=14)

# Set the title
fig.suptitle("Analysis of the impact of features expressing size on the size of benign and malignant cancer cells", fontsize=16)

plt.show()
```

## CODING TASK #6. PLOT PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS USING SEABORN

- **Plot the pairplot**
- **Strong correlation between the mean radius and mean perimeter, mean area and mean primeter**
- **plot the distplot** 
- **Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)**
- **KDE is used to plot the Probability Density of a continuous variable.** 

```python
custom_palette = sns.color_palette(["darkblue", "pink"])


sns.pairplot(
    cancer_df,
    vars=["mean area", "mean radius", "mean smoothness", "mean texture"],
    corner=True,
    hue="target",
    palette=custom_palette
)
plt.show()
```

```python
cancer_df.describe().T
```

```python
numeric_tips = cancer_df[[ "mean radius", "worst radius"]].select_dtypes(include=[np.number])

correlation_matrix = numeric_tips.corr()

sns.heatmap(correlation_matrix, annot=True);
```

```python
numeric_tips = cancer_df[[ "mean texture", "mean smoothness"]].select_dtypes(include=[np.number])

correlation_matrix = numeric_tips.corr()

sns.heatmap(correlation_matrix, annot=True);
```
